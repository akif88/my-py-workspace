```python
import pandas as pd

filepath = 'data/Orange_Telecom_Churn_Data.csv'
data = pd.read_csv(filepath)
```

```python
data.head(10)
```

```python
data.head(10).T  # transpose 
```

```python
data.columns
```

```python
data.drop(['state', 'area_code', 'phone_number'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

for col in ['intl_plan', 'voice_mail_plan', 'churned']:
    data[col] = lb.fit_transform(data[col])
```

```python
data.head(10).T
```

```python
# Mute the sklearn warning
import warnings
warnings.filterwarnings('ignore', module='sklearn')

from sklearn.preprocessing import MinMaxScaler

msc = MinMaxScaler()
```

```python
data_msc = pd.DataFrame(msc.fit_transform(data), columns=data.columns)
```

```python
data_msc.head(10).T
```

```python
data_msc.sample(n=10, replace=False).T
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data_msc, test_size=0.3) # split train, test data randomly

X_data = train_data.copy()
y_data = X_data.pop('churned')

X_Test = test_data.copy()
y_test = X_Test.pop('churned')
```

```python
X_data.T
```

```python
y_data
```

```python
X_Test.T
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_Test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_pred.round(), y_test)
acc
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()

LR = LR.fit(X_data, y_data)
y_pred = LR.predict(X_Test)
```

```python
acc = accuracy_score(y_pred.round(), y_test)
acc
```

```python
from sklearn.tree import DecisionTreeClassifier

LR = DecisionTreeClassifier()

LR = LR.fit(X_data, y_data)
y_pred = LR.predict(X_Test)
acc = accuracy_score(y_pred.round(), y_test)
acc
```
