```python
import pandas as pd
```

```python
file_1 = 'BX-Books.csv'
file_2 = 'BX-Users.csv'
file_3 = 'BX-Book-Ratings-Train.csv'
file_4 = 'BXBookRatingsTest.csv'
```

```python
data_book = pd.read_csv(file_1,  error_bad_lines=False, encoding='latin1', sep=';')
data_book.head(10)
```

```python
data_users = pd.read_csv(file_2, error_bad_lines=False, encoding='latin1', sep=';')
data_users.head(10)
```

```python
data_train = pd.read_csv(file_3,error_bad_lines=False, encoding='latin1', sep=';')
data_train.sort_values('User-ID')
```

```python
data_test = pd.read_csv(file_4,error_bad_lines=False, encoding='latin1', sep=';')
data_test.sort_values('Id')
```

```python
# don't use axis=1
data_users.dropna(inplace=True) # axis : {0 or ‘index’, 1 or ‘columns’}, default 0 Determine if rows or columns which contain missing values are removed.
```

```python
data_users.head(10)
```

```python
data_book.drop(columns=['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], inplace=True)
data_book.head(10)
```

```python
M_user_train=pd.merge(data_users, data_train, how='inner', on='User-ID') # inner for matching row value in USER-ID
M_user_train.head(10)
```

```python
M_user_book_train = pd.merge(M_user_train, data_book, how='inner', on='ISBN')
M_user_book_train.head(10)
```

```python
M_user_book_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
M_user_book_train.ISBN = LE.fit_transform(M_user_book_train.ISBN)
M_user_book_train.Location = LE.fit_transform(M_user_book_train.Location)
M_user_book_train['Book-Title'] = LE.fit_transform(M_user_book_train['Book-Title'])
M_user_book_train['Book-Author'] = LE.fit_transform(M_user_book_train['Book-Author'].astype(str))
M_user_book_train['Year-Of-Publication'] = LE.fit_transform(M_user_book_train['Year-Of-Publication'].astype(str))
M_user_book_train['Publisher'] = LE.fit_transform(M_user_book_train['Publisher'].astype(str))
```

```python
M_user_book_train.head(10)
```

```python
# Mute the sklearn warning
import warnings
warnings.filterwarnings('ignore', module='sklearn')

from sklearn.preprocessing import MinMaxScaler

y_data = M_user_book_train.pop('Book-Rating')
msc = MinMaxScaler()
data_msc = pd.DataFrame(msc.fit_transform(M_user_book_train), columns=M_user_book_train.columns)
data_msc.head(10)
```

```python
y_data.head(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_data = M_user_book_train.copy()
knn = KNeighborsClassifier(n_neighbors=10)
knn = knn.fit(X_data, y_data)
```

```python
data_test.head(10)
```

```python
data_test.rename(columns={'Id':'User-ID'}, inplace=True)
```

```python
M_user_test = pd.merge(data_users, data_test, how='inner', on='User-ID')
M_user_test
```

```python
M_user_book_test = pd.merge(data_book, M_user_test, how='inner', on='ISBN')
M_user_book_test
```

```python
M_user_book_test.ISBN = LE.fit_transform(M_user_book_test.ISBN)
M_user_book_test.Location = LE.fit_transform(M_user_book_test.Location)
M_user_book_test['Book-Title'] = LE.fit_transform(M_user_book_test['Book-Title'])
M_user_book_test['Book-Author'] = LE.fit_transform(M_user_book_test['Book-Author'].astype(str))
M_user_book_test['Year-Of-Publication'] = LE.fit_transform(M_user_book_test['Year-Of-Publication'].astype(str))
M_user_book_test['Publisher'] = LE.fit_transform(M_user_book_test['Publisher'].astype(str))
```

```python
M_user_book_test.head(10)
```

```python
y_expect = M_user_book_test.pop('Expected')
data_msc_test = pd.DataFrame(msc.fit_transform(M_user_book_test), columns=M_user_book_test.columns)
data_msc_test.head(10)
```

```python
y_expect
```

```python
X_test = M_user_book_test.copy()
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_pred, y_expect)
acc
```
